```python
import random
```

```python
class Lane:
    def __init__(self, size):
        self.size = size
        self.jobs = []
        self.remaining = self.size
        
    def add_job(self, job):
        self.jobs.append(job)
        self.remaining -= job
        
    def __eq__(self,other):
        return self.remaining == other.remaining
    
    def __lt__(self,other):
        return self.remaining < other.remaining
    
    def __gt__(self,other):
        return self.remaining > other.remaining
        
    def __str__(self):
        return '(jobs=%s, remaining=%s)' % (str(self.jobs), self.remaining)
```

```python
class laneList(object):
    def __init__(self, lane_size,k):
        self.lane_size = lane_size
        self.lanes = []
        self.numLanes = 0
        self.k = k
        self.extra = 0
        
    def add_lane(self):
        self.lanes.append(Lane(self.lane_size))
        self.numLanes = len(self.lanes)
        self.lanes = sorted(self.lanes)
        
    def get_extra(self):
        self.num_extra = 0
        for i in range(self.numLanes):
            self.num_extra += self.lanes[i].remaining
    
    def BF(self, job):
        if self.numLanes == 0:
            self.add_lane()
        _minFit = self.lane_size + 1
        for lane in self.lanes:
            rem = lane.remaining - job 
            if rem >= 0 and rem < _minFit:
                _minFit = rem
                _bestLane = lane

        if _minFit < self.lane_size + 1:
            _bestLane.add_job(job)
  
        else:
            self.add_lane()
            self.lanes[self.numLanes-1].add_job(job)

        self.lanes = sorted(self.lanes)
        self.get_extra()
        
    def FF(self, job):
        if self.numLanes == 0:
            self.add_lane()
        notDone = True
        _curLane = 0
        while notDone:
            # print(_curLane)
            lane = self.lanes[_curLane]
            if lane.remaining - job >= 0:
                lane.add_job(job)
                notDone = False
                
            else:
                if _curLane + 1 <= self.numLanes - 1:
                    _curLane += 1
                else:
                    self.add_lane()
            self.lanes = sorted(self.lanes)
            self.get_extra()

    
    def __str__(self):
        if self.numLanes != 0:
            for lane in self.lanes:
                print(lane)
            return 'Number of Lanes: %s, Free Space: %s' %(self.numLanes, self.num_extra)
        else: 
            return 'No Lanes'
```

```python
jobList = [random.randint(1,100) for i in range(500)]
jobList1 = sorted(jobList)
jobList2 = sorted(jobList, reverse=True)
```

```python
lanes = laneList(150,4)
for job in jobList2:
    lanes.FF(job)
print(lanes)
```

```python
lanes = laneList(150,4)
for job in jobList2:
    lanes.BF(job)
print(lanes)
```
